On July 14, 2015, an anti-abortion group released an online video that members said documented Planned Parenthood selling fetal organs. The group also claimed the video proved that Planned Parenthood, one of the most established abortion providers in the United States, altered normal abortion procedures so as to preserve the organs (CNN, 2015).

The video and the outcry it unleashed created what one advocacy group called a "media frenzy" around abortion. In a matter of days, anti-abortion groups, Republicans and the media ratcheted up the already searing debate around one of the most contested issues of our time: abortion (Sisson and Kimbert, 2014). The procedure has long been contentious in part because it is one of the few practices that so firmly straddles "the problematic intersection of public policy and private life." Few other decisions involving such a  deeply personal choice intersect with multiple spheres of public life in the way terminating a pregnancy does: health care, political policy and religion, as well as competing cultural values and narratives (Sisson and Kimbert, 2014).

In the middle are women. They are most likely to be affected by laws regulating abortion. Yet they have largely been marginalized in the public debate over abortion, which has been fought in institutional arenas, most significantly the courts, congress and the media (Feltham-King and Macleod, 2015). 

To more closely examine the dynamics of women's representation around abortion, I selected coverage of abortion by the New York Times between July 1 and Aug. 1, 2015, approximately two weeks before and two weeks after the video involving Planned Parenthood was released. I also looked at tweets during that time using the hashtag #abortion. LINK TO GitHUB

I chose the news media because it is one of the most influential spaces where the dispute over who should legitimately control abortion takes place; under what circumstances it should be permitted, if at all; and who and where it should be practiced (Feltham-King and Macleod, 2015).

Digital communication, particularly social media, has been credited with undercutting that dynamic by giving individuals a tool to bypass the gatekeepers, meaning the established press. It is a space "where online users transform the dynamics of publishing and distributing information" (Bonini and Sellas, 2014). Protest groups in particular learned to take advantage of social media to override layers of state, religious and cultural power (Shi, 2013), including the news media although the extent to which this translated into change is still being questioned (Cadena et al, 2015; Dubois and Gaffney, 2014; González-Bailón, Kaltenbrunner and Banchs, 2010; González-Bailón, Borge-Holthoefer and Moreno, 2013; Lycarião, 2015). 

But they have shown the potential to increase public attention to discourses regarding sex and gender thus mobilizing resources for the promotion of gender equity, and strengthening feminist networks (Edwards, 2004). Contrasting the expression by women on social media with reporting by news outlets provides the potential for a rich examination of the extent to which the news media is mobilized as a public authority to enforce/reaffirm cultural norms and as an agent of social control (Kraidy and Mourad, 2014), as well as how women resist that control by using alternative media. 

Existing research examining abortion coverage and social media tend to focus on news outlets and abortion activist organizations rather than individual women specifically. As one advocate put it, "We hear about abortion but we never hear a woman’s story behind it" (Himiak, L, personal communication, October 26, 2015). 

This investigation adds to the literature by pulling those less examined elements together. The paper looks at the difference between how women used media to communicate directly about abortion and how their representation differed from the news coverage about abortion. By examining news media and social media this article tries to shed new light on the media-state-society relations within the context of reproductive politics and whether social media has allowed women to resist state control of reproduction (Shi, 2014) in new ways that other forms of media do not support or work against. 

In other words, this paper asks what it sounds like when women speak for themselves (Jones and Jerman, 2014). 

If control over reproduction is a measure of empowerment (Upadhyay et al, 2014), these questions are important for women. But they are also important for news sites to understand. The New York Times and other news organizations have acknowledged that they have excluded women online and offline (Sullivan, 2015; Tessier, 2015). Women represent an untapped and much-needed market for the papers that they are eager to capture (Tessier, 2015). In addition to being a problem for their bottom lines, the absence of women in news sites is a problem for democracy (Tessier, 2015).

###Discussion

Public opinion about abortion has long fallen along a spectrum that fluctuates and rarely aligns as simply as "anti" or "pro" abortion (Ziegler, 2015). The issue traverses a multifaceted history of attitudes toward women based on race, ethnicity, class, religion, changing sex roles (Ziegler, 2015), national identity (Kraidy and Mourad, 2014) and the source of political authority (Hampton, 1997). All of these have changed over time and are still in flux nearly 43 years since Roe v. Wade, making reductive understandings about abortion commonplace and easy. (Ziegler, 2015) But it is also true that two institutional sides solidified after the Supreme Court decision and that, even though the music has changed over the years, they have been partners locked in the same dance steps ever since. 

Social media, however, is new. And it provides an avenue for women to tell their story publicly and impact the political process (González-Bailón, Banchs and Kaltenbrunner, 2012). 

The question here is how women engage in and are engaged by those platforms around abortion discourse: How what they say about abortion challenges frames and rhetoric constructed by the news media and, by extension, the state's attempts to assert control over women's bodies and reproduction. 

Not surprisingly, the picture that emerges is, like the issue, multifaceted. 

The first online story about the video that members said documented Planned Parenthood selling fetal organs appeared shortly after its July 14 release. The New York Times ran the first print story about the video the next day. 

Before that, during the two weeks ending on July 14, the paper ran a dozen news stories with the word abortion in the content.  The number climbed to 22 after July 15.  Fifteen of those carried bylines by women reporters, 19 had bylines by men and one carried no byline because it was a roundup of news with only some bylined. More than half of the stories (18) focused on the upcoming presidential election and Republicans dominated 19 of them. 

Women as sources of authority were quoted in 24 of the stories; some pieces had multiple women authority sources while 11 lacked any women at all. Otherwise, the reactions cited came from "conservatives" including Heritage for America, as well as presidential hopefuls governors Bobby Jindal, Rick Perry and Greg Abbott with few reactions from Democrats. 

Only two stories quoted a woman who did not operate in an official capacity, that is, a "regular" woman who would be directly affected by the policies being fought for by the Republicans overwhelmingly represented in the sample.  However, the two stories that quoted "regular" women were about a shooting rampage. The women quoted were friends of the victim, who was an anti-abortion activist.  

By far the most noticeable context for abortion, before and after the video was released, was in coverage of Republican candidates and congressional attempts to limit the procedures. The coverage did not steer away in a significant way from its political focus before or after the release of the videos, which began in print on July 15 with "Video Intensifies Abortion Debate" (Calmes, 2015) quoting a Planned Parenthood spokesman. 

###Social Media 

Women mainly used Twitter in four ways. They referenced news stories; re-broadcast political messages; expressed their opinions about abortion; and interacted with campaigns aimed at persuading public opinion. This was true whether they were using Twitter to oppose or to support abortion access. 

Few differences were found in the first two categories between how women used social media to talk about abortion compared to the news media coverage. In both categories, women were sharing material that they had little or no input in creating and doing so in order to underscore their support for or opposition against abortion access, or possibly to establish social status (Journal of Pragmatics, 2014). 

But they frequently were doing so in bolder language than The New York Times would have permitted (Diamond, 2012). 

###Framing

Tweets organized by a campaign diverged the most significantly, particularly around framing.

Framing is a tool reporters use to condense and package information into recognizable content that the public can easily understand. The information fits into culturally constructed and reinforced cognitive schema recognized by the public and defined by cultural elites. (Gitlin, 1980; Entman, 2007 ).

Common frames pit two sides against each other, use battle analogies, evoke emotion and danger. In the Times coverage, for example, reporters emphasized the gamesmanship of the man who released the video and the Republican response to the content. Reporters used words like "battle," "war," "dispute," "offensive." 

In one story, "Democrats and abortion rights supporters are bracing for a sustained sophisticated and coordinated effort to force a debate on the uncomfortable moral and ethical questions that abortion raises" (Peters, 2015).  

As the stakes for women rose with the escalation of Republican attempts to block federal funding for Planned Parenthood, the news coverage zoomed in even closer on the political wargames. Meanwhile, women took to social media to stake their claim to the public debate (as did Planned Parenthood and NRLC). Twitter allowed the women who support abortion access to object to their silence and to reject the treatment of abortion as a political game. 

Abortion risks and emotional toll also are common codes used in news coverage. This device allows reporters to frame the availability or unavailability of abortion in a way that highlights the risks, dangers or emotional toll for women (Elmore, 2008) and implicitly associates abortion with subversion. 

Again, the reporter quoted above, Jeremy Peters, writes that the Republicans will "force a debate on the uncomfortable moral and ethical questions that abortion raises."  The wording positions opposition to abortion as the normative standard in the issue and shapes the debate whether the reporter was aware or not of the ideological code at play (Kleinman and Ezzell, 2012). 

We can see this dynamic clearly in an article about the #ShoutYourAbortion social media campaign, which grew out of a response to the Republican threat to freeze federal funding to Planned Parenthood because of the fetal tissue charges. 

 The article is the first mention in the Times of the campaign, then two weeks old, originated by two women, Amelia Bonow and Lindy West.  

The article is outside the original sample but it is worth including in this examination because it is one of the very few Times articles to directly references "regular" women and their use of social media with regard to abortion, and because the campaign played a part in how women used Twitter regarding abortion messages. 

The reporter who wrote the story, Tamar Lewin, explains that the goal of the #ShoutYourAbortion campaign was to "encourage women who had kept their abortions secret to speak up — in an effort to reframe the debate on the subject." This is in keeping with the campaign's explicitly articulated goal of  destigmatizing abortion in order to weaken attacks on access. “Having an abortion made me happy in a totally unqualified way. Why wouldn’t I be happy that I was not forced to become a mother?" Bonow wrote in her initial Facebook post. 

But Lewin proceeded to focus exactly on angles that the campaign rejected by its attempt to normalize abortion and decrease the shame and controversy used by opponents to attack access to the procedure. 

For example, Lewin wrote that, "Many of the Twitter messages showed the emotion surrounding the decision to have an abortion. Some women wrote about being pregnant and scared at 15, or living with an abusive man who might harm a child." 

It is true that women experience fear and stigma around ending a pregnancy but it varies in association with religious affilation, race, ethnicity age, education and motherhood status (Cockrill, Upadhyay, Turan and Greene Foster, 2013)  Yet, the focus of the story first details fear in order to legitimate abortion. Then Lewin moves on to the conflict around the posts and how Republicans had put Planned Parenthood "on the defensive."  She highlights the backlash against the campaign, as well as against Bonow and West.  Once those frames have been set, Lewin's reporting switches to the nature of the campaign, which she characterizes as bringing abortion "out of the shadows." 

"It has been very rare for women to come out in public about an abortion," Lewin writes. She then described Jackie Spiers' announcement on the House floor that she had an abortion. This would discount the content in blogs, podcasts, social media and alternative news media, as well as decades of protests, in which women discuss abortion openly. 

Lewin then turns to another frequent frame used by reporters about political protests, that of trivializing, marginalizing and casting doubt on the campaign's effectiveness particularly by discounting its numbers (Gitlin, 1980). 
 
"#ShoutYourAbortion posts dwindled to a couple thousand a day, with fewer people sharing their stories and the chain devolving into both sides arguing the usual points on the issues," Lewin wrote in the penultimate graph. In fact, Bonow and West launched a fundraiser and had turned to other forms of media to try to sustain the momentum of the social media attention. 

That does not mean the campaign was a success. In fact, as I point out later, a gap exists between campaign and effectiveness.  But the angles Lewin chose contrasted sharply with the women's #ShoutYourAbortion posts and the narrative they tried to create. 

###Pro-life 

Anti-abortion advocates also complain about the news media although they tend to talk about it in terms of a liberal bias (Armstrong and Boyle, 2011; Carmines, Gerrity and Wagner, 2010; Douthat, 2012; Shaw, 1990; Shellnut, 2015; Swallow-Prior, 2013; ). Indeed, women may see an even wider gap between the coverage and their construction of abortion if they are religious and believe abortion is wrong because it interferes with God's will (Press and Cole, 1999; Shellnut, 2015; Swallow-Prior, 2013). 

Because the perspective of religiously conservative anti-abortion women is in opposition to most mainstream reporting, they turn to alternative authorities to legitimate their perspectives (Press and Cole, 1999). These are trusted sources that anti-abortion access women retweeted and quoted in their social media posts. These resources, in turn, contained information in step with anti-abortion campaigns. 

That said, pro-life organizations like NRLC, one of the nation's oldest, have been slow to take advantage of social media and made a point of catching up (Andrusko, 2009; Dobson, 2009; Little, 2008; McGee, 2008). In general, anti-access groups have recognized the potential to create campaigns, like this one by a site called Life Site News, which made specific reference to "baby parts", retweeted by women and men:

###Women's Own Words 

Otherwise, individual, independent messages tend to be muted without the help of a hashtag shared among networks. One exception came from the opinion category, in particular pro-access tweets involving the demand for autonomy. 

A potential for overlap would seem to exist between the news media and women around this perspective because it is a common theme in news coverage (Elmore, 2008). 

But it was unclear how much the claim to autonomy is a reflection of messaging from pro-abortion access leaders, which the women may have been repeating.  The overlap then could have been explained because the news media relies on these organizations for quotes. So both women and the news media would be echoing the messaging. 

This constitutes an example of the news media and social media bringing a "manufactured public world into private space" (Gitlin, 1980). Both have become "core systems for the distribution of ideology" (Gitlin, 1980). 

The difference is that social media offers a space where individuals can contest those systems and the agencies of control that operate within them and perpetuate political and economic elite definitions of reality (Gitlin, 1980). 

###Conclusion 

Social media allows individuals to bypass the news media gatekeepers but being visible is another story, which shows the strength of organized campaigns in an online context. The messages conveyed in the women's tweets were most successful when there was an organized campaign behind the hashtags. 
 
This is consistent with previous research that the internet and social media and other internet technology are useful for helping individuals to activate mobilization but that they have not changed the way the media system works. Lycarião (2015) calls this lack of structural transformation "old wine in a new bottle" (p. 2).  

His comment is meant to say that the basic structures of political communication cycle today remains similar to the traditional mass communication era. Significantly, the news media still control the instruments for the production of visibility and making the public agenda discernible, which is a function of the social context to which media systems are connected (Lycarião, 2015; Yardi & Boyd, 2010; Fatkin and Lansdown, 2015).  This, he claims, is because visibility is still focused on the political elites and their status varies much more with socio-political structures than with the internet's ability to broadcast information. Lycarião also invokes a reminder that technology is an extension of the humans who create it; that it is "'co-constrcuted' by the the social forces it organizes and unleashes" (Feinberg 2011 as cited in Lycarião, 2015, p. 10). 

Thus social media alone are not effective agents of change except in rare circumstances.  They are tools to raise awareness and share information among a wide network of people who use social media. They can be leveraged by groups to strategically make an issue visible to the news media or to challenge frames perpetuated in the media, as well as convey public opinion. And they can be instrumental in mass mobilization. 

Women can take advantage of social media to make their voice heard by forming a "vibrant and increasingly visible collaborative force," that widens the public sphere (Diamond, 2012, p. 5 ).  To borrow from Howard and Hussein's analysis, they can use social media to help turn individualized, localized and community-specific dissent into a structured movement with a collective consciousness about both shared plights and opportunities for action. #ActuallyWomen, #standwithwendy and #ShoutYourAbortion (as well as #BlackLivesMatter, #standwithwendy, #BlackTwitter) showed that campaigns can be effective in sparking this process. The same holds true for anti-abortion campaigns. But Twitter campaigns have also shown that to become visible beyond the sphere of a defined network they often still have to rely on the news media to broadcast their message if they want broad-based attention.

An interesting question for the future will be how networks can override the logic of the media system in order to influence policy decisions (Rucht, 2004; Simon and Jerit, 2007). In the meantime, relying on the news media for visibility presents a dilemma for women and for protest groups overall. By engaging the media they risk losing control of their message. Reporters use socially accepted frames and rhetorical norms to choose specific information and organize experience (Gitlin, 1980). Those filters are most often based on the values of socially dominant groups, making the news media a powerful tool for elites to influence the distribution of power: who gets what, when and how (Entman, 2007).

Currently news organizations are struggling to create effective online environments that attract a larger audience, in turn driving advertising. But the sample examined here shows that they are falling back on the practices that marginalize groups like women and protesters and that drive these groups away from the news media. The news media's reliance on Republican abortion-access opponents helped shape the public perception of abortion in a way that excludes women and simplifies a multidimensional issue to the detriment of women and the public. 

Journalists can change this dynamic by using social media to find "regular" women to include in their reporting, helping to establish better connections with a variety of sources, rather than politicians. But they will have to be self-reflexive in their reporting to avoid alienating women, whether they are pro-life, pro-abortion or somewhere along the spectrum. This can help journalism fulfill its mission of informing the public, and help news outlets access the market women represent. 

Otherwise, efforts to tap their financial potential are likely to fail and the public sphere that journalism helps to create will continue to miss half its citizens. In the process, the voices of real women are lost and a divisive debate dominates. 

###Limitations and next steps 
Given the small sample used here, this research be interpreted with caution. However, some of the patterns that emerged are supported by previous research, including framing and the dynamics of the media system and organizing. That said, more research using a larger sample is needed to give a deeper understanding of how news media coverage and social media have influenced the conversation around abortion specifically and whether either has had consequences for the success of legislation or influencing public opinion.
